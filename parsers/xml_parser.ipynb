```python
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
tree = ET.parse('../data/PointzAggregator-AirlinesData.xml')
```

```python
root = tree.getroot()
```

```python
columns = ['uid', 'first_name', 'last_name', 'cards_type', 'card_number', 'bonusprogram', 'code', 'date', 'departure', 'arrival', 'fare']
```

```python
data = []

users = root.findall('user')

index = 0
for user in users:
    uid = first = last = type = number = bonusprogramm = None
    uid  = user.get('uid')
    name = user.find('name')
    if name is not None:
        first = name.get('first')
        last  = name.get('last')
        
    cards = user.find('cards')
    if cards is not None:
        type = cards.get('type')
        
        card = cards.find('card')
        if card is not None:
            number = card.get('number')

            bonusprogramm_element = card.find('bonusprogramm')
            if bonusprogramm_element is not None:
                bonusprogramm = bonusprogramm_element.text

            activities = card.find('activities')
            if (activities is not None):
                for activity in activities.findall('activity'):
                    code = date = departure = arrival = fare = None
                    
                    code_element = activity.find('Code')
                    if (code_element is not None):
                        code = code_element.text
                        
                    date_element = activity.find('Date')
                    if (date_element is not None):
                        date = date_element.text

                    departure_element = activity.find('Departure')
                    if (departure_element is not None):
                        departure = departure_element.text

                    arrival_element = activity.find('Arrival')
                    if (arrival_element is not None):
                        arrival = arrival_element.text

                    fare_element = activity.find('Fare')
                    if (fare_element is not None):
                        fare = fare_element.text

                    data.append([uid, first, last, type, number, bonusprogramm, code, date, departure, arrival, fare])

    
```

```python
df = pd.DataFrame(data, columns=columns)
```

```python
df.to_csv('../data/PointzAggregator-AirlinesData.csv', index_label='id')
```
