```python
import pandas as pd
import pickle
```

```python
file_path = '../data/airoport_data.pkl'
with open(file_path, 'rb') as file:
    countries = pickle.load(file)

file_path = '../data/countries_continents.pkl'
with open(file_path, 'rb') as file:
    continents = pickle.load(file)
```

```python
sky = pd.read_csv('../data/SkyTeam-Exchange.csv', index_col='id')
```

### Отсутствуют имена для идентификации пассажиров, поэтому добавим таблицу с номерами карт и никнеймами

```python
pointz = pd.read_csv('../data/PointzAggregator-AirlinesData.csv', index_col='id')
```

```python
pointz
```

```python
sky.rename(columns={'BoardNum':'CardNumber'}, inplace=True)
```

```python
pointz['name'] = pointz['FirstName'] + ' ' + pointz['LastName']
```

```python
card_numbers = pointz[['uid', 'name', 'CardNumber']]
```

```python
card_number = card_numbers.drop_duplicates()
card_number
```

### Соединяем таблицы по номеру карт

```python
full_sky = pd.merge(sky, card_numbers, how='left', on='CardNumber')
```

```python
full_sky.drop_duplicates(inplace=True)
full_sky
```

### Избавимся от пассажров, которых невозможно идентифицировать

```python
full_sky.isna().sum()
```

```python
full_sky.dropna(inplace=True)
```

```python
full_sky.isna().sum()
```

### Переведем именна колонок в snake_case

```python
full_sky.columns = [
    'data',
    'flight_num',
    'card_number',
    'class',
    'fare',
    'from',
    'status',
    'to',
    'uid',
    'name',
]
```

```python
full_sky
```

### Определим страны по коду аэропорта

```python
full_sky['departure_country'] = full_sky['from'].map(countries)
full_sky['arrival_country']   = full_sky['to'].map(countries)
```

### Определим континенты по странам

```python
full_sky['departure_continent'] = full_sky['departure_country'].map(continents).fillna('Asia')
full_sky['arrival_continent']   = full_sky['arrival_country'].map(continents).fillna('Asia')
```

### Создаем таблицу данных

```python
def count_suspicious_flights(df):
    prev_airport = None
    susp_flights = cross_country = susp_cross_country = long_flights = 0
    for src_airport, dst_airport, src_country, dst_country, src_continent, dst_continent in df.values:
        if prev_airport != None and src_airport.lower() != prev_airport.lower():
            susp_flights += 1
            if src_country.lower() != dst_country.lower():
                susp_cross_country += 1
        prev_airport = dst_airport
        
        if src_country.lower() != dst_country.lower():
            cross_country += 1
        if src_continent.lower() != dst_continent.lower():
            long_flights += 1
    return susp_flights, cross_country, susp_cross_country, long_flights
```

```python
suspicious_flights_df = pd.DataFrame(columns=['name', 'flights_count', 'suspicious_flights', 
                                              'cross_country', 'suspicious_cross_country', 'long_flights'])

for i, uid in enumerate(full_sky.uid.unique()):
    df = full_sky[full_sky['uid'] == uid].sort_values(by='data').reset_index(drop=True)
    name = full_sky['name'].iloc[i]
    flights_count = len(df)
    
    susp_flights, cross_country, susp_cross_country, long_flights = count_suspicious_flights(df[['from', 'to', 
                                                                                                 'departure_country', 'arrival_country', 
                                                                                                 'departure_continent', 'arrival_continent']])
    
    row_to_add = [name, flights_count, susp_flights, cross_country, susp_cross_country, long_flights]
    suspicious_flights_df.loc[i] = row_to_add
```

### Вычислим процент подозрительных и международных рейсов

```python
suspicious_flights_df['percent_suspicious_flights'] = suspicious_flights_df['suspicious_flights'] / suspicious_flights_df['flights_count']
```

```python
suspicious_flights_df['percent_cross_country'] = suspicious_flights_df['cross_country'] / suspicious_flights_df['flights_count']
```
